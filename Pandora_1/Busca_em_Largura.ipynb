## Obtendo a lista de SUB

```python
ctmt.head(3)
```

```python
dict_SUB = {}
for idx, row in ctmt.iterrows():
    cod_ctmt = row['COD_ID']
    sub = row['SUB']
    descr = row['DESCR']
    if sub not in dict_SUB:
        dict_SUB[sub] = {}
        dict_SUB[sub]['cod_ctmt'] = cod_ctmt
        dict_SUB[sub]['nome'] = descr
dict_SUB
```

**Terminou esse bloco com `dict_SUB` com chaves sendo os códigos de SUB.**

## Obtendo a lista de PAC's dos circuitos que se originam em cada SUB

```python
ctmt.head(3)
```

```python
for sub in dict_SUB:
    flag = (sub == ctmt['SUB'])
    ctmt_filter = ctmt[flag]
    dict_SUB[sub]['pac_circuitos'] = list(ctmt_filter['PAC'])
dict_SUB
```

## Obtendo os trechos que formam os circuitos

Os trechos estão em `ssdmt`.

```python
ssdmt.head(3)
```

Vamos criar colunas novas para registrar subestação, circuito, de, para e contador.

```python
ssdmt_com_circ = ssdmt.copy()
ssdmt_com_circ['circ_SUB'] = -1
ssdmt_com_circ['circ_CTMT'] = -1
ssdmt_com_circ['circ_PAC_de'] = -1
ssdmt_com_circ['circ_PAC_pr'] = -1
ssdmt_com_circ['circ_cont'] = -1
ssdmt_com_circ.head(3)
```

Rotina que faz a busca pelos trechos que pertencem a certo circuito.

```python
def busca_em_largura(df, sub, pac_circ):
    flag = (df['circ_SUB'] == -1)
    dff = df[flag]
    
    C = [pac_circ,]
    D = set(list(dff['PAC_1']) + list(dff['PAC_2'])) - set(C)
    cont = 0

    while(len(C) > 0):

        atual = C[0]
#         print('\n\n')
#         print(f'atual = {atual}\n')
#         print('=====')
#         for item in C:
#             print(f'C.item = {item}')
#         print('=====')

        flag1 = (dff['PAC_1'] == atual)
#         print(f'sum(flag1) = {sum(flag1)}\n')

        filhos = dff[flag1]
#         display(filhos)

        for idx, row in filhos.iterrows():
            if row['PAC_2'] in D:
                print('Achou trecho!')
                df.loc[idx,'circ_SUB'] = sub
                df.loc[idx,'circ_CTMT'] = pac_circ
                df.loc[idx,'circ_PAC_de'] = row['PAC_1']
                df.loc[idx,'circ_PAC_pr'] = row['PAC_2']
                df.loc[idx,'circ_cont'] = cont
                cont += 1

                D.remove(row['PAC_2'])
                C.append(row['PAC_2'])
            else:
                print('Achou malha!')
                df.loc[idx,'circ_SUB'] = 'Malha'
                df.loc[idx,'circ_CTMT'] = 'Malha'
                df.loc[idx,'circ_PAC_de'] = 'Malha'
                df.loc[idx,'circ_PAC_pr'] = 'Malha'

        flag2 = (dff['PAC_2'] == atual)
#         print(f'sum(flag2) = {sum(flag2)}\n')
        
        filhos = dff[flag2]
#         display(filhos)
        
        for idx, row in filhos.iterrows():
            if row['PAC_1'] in D:
                print('Achou trecho!')
                df.loc[idx,'circ_SUB'] = sub
                df.loc[idx,'circ_CTMT'] = pac_circ
                df.loc[idx,'circ_PAC_de'] = row['PAC_2']
                df.loc[idx,'circ_PAC_pr'] = row['PAC_1']
                df.loc[idx,'circ_cont'] = cont
                cont += 1

                D.remove(row['PAC_1'])
                C.append(row['PAC_1'])
            else:
                print('Achou malha!')
                df.loc[idx,'circ_SUB'] = 'Malha'
                df.loc[idx,'circ_CTMT'] = 'Malha'
                df.loc[idx,'circ_PAC_de'] = 'Malha'
                df.loc[idx,'circ_PAC_pr'] = 'Malha'
    
        flag = (df['circ_SUB'] == -1)
#         print(f'sum(flag = {sum(flag)}')
        
        dff = df[flag]
        C.remove(atual)
```

Fazendo para apenas a 1° subestação e o 1° circuito.

```python
df = ssdmt_com_circ.copy()
sub = next(iter(dict_SUB))
circ = dict_SUB[sub]['pac_circuitos'][0]
busca_em_largura(df, sub, circ)

flag = (df['circ_SUB'] != -1)
print(f'df[flag].shape = {df[flag].shape}')
print(f'df[~flag].shape = {df[~flag].shape}')
df[flag]
```
